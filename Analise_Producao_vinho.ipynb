```python
import pandas as pd
```

```python
dados_producao = pd.read_csv ("/content/Producao.csv", encoding="UTF-8-SIG",sep=";",thousands=".",decimal=",")
```

```python
dados_producao.head()
```

```python
dados_producao["Categoria"] = dados_producao["id"]
dados_producao["Nome Produto"] = dados_producao["produto"]
dados_producao = dados_producao.set_index('id')
```

```python
dados_producao.head()
```

```python
dados_producao= dados_producao[["Nome Produto",'Categoria', '2009', '2010', '2011', '2012',
       '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021',
       '2022', '2023']]
```

```python
dados_producao
```

```python
dados_producao["Categoria"] = dados_producao["Categoria"].mask(dados_producao["Categoria"] >= 15,'derivado')
```

```python
dados_producao.head()
```

```python
dados_producao["Categoria"] = dados_producao["Categoria"].replace([5,6,7,8],"vinho fino vinífera")
dados_producao["Categoria"] = dados_producao["Categoria"].replace([9, 10, 11, 12, 13, 14],"suco")
dados_producao["Categoria"] = dados_producao["Categoria"].replace([1,2,3,4],"vinho mesa")
```

```python
dados_producao
```

```python
dados_producao = dados_producao.drop(1, axis=0)
dados_producao = dados_producao.drop(5, axis=0)
dados_producao = dados_producao.drop(9, axis=0)
dados_producao = dados_producao.drop(15, axis=0)
```

```python
dados_producao
```

```python
dados_producao = pd.melt(dados_producao, id_vars=["Categoria","Nome Produto"],var_name="Ano", value_name="Total_Produzido")
dados_producao["Nome Produto"] = dados_producao["Nome Produto"].str.lower()
```

```python
dados_producao.head(15)
```

```python
print(f"# linhas da tabela: {dados_producao.shape[0]}")
print(f"# linhas da tabela: {dados_producao.shape[1]}")
print(f"Min ano da tabela: {dados_producao['Ano'].min()}")
print(f"Max ano da tabela : {dados_producao['Ano'].max()}")
```

```python
dados_producao = dados_producao.drop_duplicates()
dados_producao = dados_producao.dropna()
```

```python
dados_producao["Total_Produzido"] = dados_producao.Total_Produzido.astype('int64')
dados_producao.info()
```
